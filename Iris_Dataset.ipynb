```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
from sklearn.model_selection import cross_val_score , cross_val_predict
from sklearn.ensemble import RandomForestClassifier

data = pd.read_csv("iris_dataset.csv")  # Downloaded the dataset from kaggle.

data.head()  # This will print the first 5 rows from the dataset.
data.tail()  # This wll print the last 5 rows from the dataset.

data.dtypes  # This will tell the data type of every column in the dataset.
print(type(data))  # DataFrame
data.shape  # Number of rows and columns in the dataset

plt.figure(figsize = [10 , 5])
sns.boxplot(x = "Species" , y = "PetalWidthCm" , data = data , hue = "Species")  # hue will provide colour.
plt.title("BOX PLOT")
plt.show()

data["Species"].unique()

data.keys()
data.values

data["SepalLengthCm"].hist()  # Histogram for the SepalLengthCm column.

sns.scatterplot(x = "PetalWidthCm" , y = "SepalWidthCm" , data = data , hue = "Species")  # Total 16 combinations can be made.

data1 = data.drop("Id" , axis = 1)
sns.pairplot(data1 , hue = "Species")  # This will print all 16 combinations at once.

x = data.drop(["Species" , "Id"] , axis = 1)
y = data["Species"]

x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2)  # Testing data is 20%, training data is 80%.

x_train.shape
x_test.shape
y_train.shape
y_test.shape

model_lr = LogisticRegression(max_iter = 1000)  # Using LogisticRegression
model_lr.fit(x_train , y_train)

y_pred = model_lr.predict(x_test)

accuracy = accuracy_score(y_test , y_pred)
np.round(accuracy , 4) * 100

report = classification_report(y_test , y_pred)
print(report)


model_logisticR = LogisticRegression(max_iter = 1000)  # Default max_iter is 100.
accuracies = cross_val_score(model_logisticR , x_train , y_train , cv = 5)
np.mean(accuracies)

y_predz = cross_val_predict(model_logisticR , x_train , y_train)  # Using Cross-Validation.
y_predz

predicted_correctly = y_predz == y_train  # Will print True for the values that were predicted correctly.
not_predicted_correctly = ~predicted_correctly  # Will print False for the values that were not predcited correctly.

predicted_correctly
not_predicted_correctly


model_rf = RandomForestClassifier()  # Using RandomForestClassifier
model_rf.fit(x_train , y_train)
y_pred = model_rf.predict(x_test)
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
labels = sorted(set(y_test))
con_matrix = confusion_matrix(y_test , y_pred)
sns.heatmap(con_matrix , annot = True , cmap = "plasma" , xticklabels = labels , yticklabels = labels)
```
